```python
import pandas as pd
import random
from translate import Translator
import os

os.system("cls" if os.name == "nt" else "clear")  # Terminal ekranını temizleme

SHEET_ID = '1xIgcG6shp1af0eggfUkI3QUycZ0v1233UEwnH6wiU8g'  # Google Sheets ID
SHEET_NAME = 'less-real_data'  # Google Sheets sayfası adı
sheet_url = 'https://docs.google.com/spreadsheets/d/1xIgcG6shp1af0eggfUkI3QUycZ0v1233UEwnH6wiU8g/edit#gid=2002820014'  # Google Sheets URL
url_1 = sheet_url.replace('/edit#gid=', '/export?format=csv&gid=')  # CSV URL

# Load data from Google Sheets into a DataFrame
df = pd.read_csv(url_1)

# A1 boş ise ilk satırı atlayalım
df = df.iloc[1:]

# Veri çerçevesinde rastgele bir indeks seçelim
random_index = random.randint(0, len(df) - 1)
random_row = df.iloc[random_index]

# Anime ismi ve karakteri yan yana, alt satırda alıntı ile birleştirelim
output = f"{random_row['Anime']} - {random_row['Character']}\n{random_row['Quote']}"

print("Original Output:\n")
print(output + "\n")

# Translate the output to Turkish
translator = Translator(to_lang="tr")
translated_output = translator.translate(output)

print("Translated Output:\n")
print(translated_output)
```
